```python
from time import time
import numpy as np
import matplotlib.pyplot as plt

tiempo=[]
n = [100,200,300,400, 500]

for i in range(len(n)):
                       
    A = np.mat(np.full((n[i], n[i]), 1, dtype=int))
    B = np.mat(np.full((n[i], n[i]), 2, dtype=int))
    tiempo_inicial = time()
    C=A*B
    tiempo_final = time()
    tiempo.append( (tiempo_final - tiempo_inicial))


    
```

```python
plt.plot(n,tiempo)
plt.xlabel("n")
plt.ylabel("tiempos")
plt.show()
```
